```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# Define color mapping for each size
size_colors = {
    7: 'yellow',
    8: 'orange',
    9: 'darkorange',
    10: 'red',
    11: 'darkred'
}

# Map colors to each row in the DataFrame
df['color'] = df['size'].map(size_colors)

plt.figure(figsize=(12, 5))

# Time comparison (log-log)
plt.subplot(1, 2, 1)
plt.scatter(df['time'], df['time_hh'], alpha=0.7, c=df['color'])
min_val_time = min(df['time'].min(), df['time_hh'].min())
max_val_time = max(df['time'].max(), df['time_hh'].max())
plt.plot([min_val_time, max_val_time], [min_val_time, max_val_time], 'r--', label='y = x')

# Count points above and below the line y = x
above_line_time = np.sum(df['time_hh'] > df['time'])
below_line_time = np.sum(df['time_hh'] < df['time'])
plt.text(0.05, 0.95, f'Above: {above_line_time}\nBelow: {below_line_time}', transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', horizontalalignment='left', fontweight='bold')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Standard DFS Time', fontsize=12, fontweight='bold')
plt.ylabel('Color DFS + FC + MCV Time', fontsize=12, fontweight='bold')
plt.title('Time Comparison', fontsize=16, fontweight='bold')
plt.grid(False)

# Steps comparison (log-log)
plt.subplot(1, 2, 2)
valid_steps = df.dropna(subset=['steps', 'steps_hh'])
plt.scatter(valid_steps['steps'], valid_steps['steps_hh'], alpha=0.7, c=valid_steps['color'])
min_val_steps = min(valid_steps['steps'].min(), valid_steps['steps_hh'].min())
max_val_steps = max(valid_steps['steps'].max(), valid_steps['steps_hh'].max())
plt.plot([min_val_steps, max_val_steps], [min_val_steps, max_val_steps], 'r--', label='y = x')

# Count points above and below the line y = x
above_line_steps = np.sum(valid_steps['steps_hh'] > valid_steps['steps'])
below_line_steps = np.sum(valid_steps['steps_hh'] < valid_steps['steps'])
plt.text(0.05, 0.95, f'Above: {above_line_steps}\nBelow: {below_line_steps}', transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', horizontalalignment='left', fontweight='bold')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Standard DFS Steps', fontsize=12, fontweight='bold')
plt.ylabel('Color DFS + FC + MCV Steps', fontsize=12, fontweight='bold')
plt.title('Steps Comparison', fontsize=16, fontweight='bold')
plt.grid(False)

# Add custom legend for sizes
legend_elements = [Patch(facecolor=color, edgecolor='k', label=f'Size {size}') for size, color in size_colors.items()]

# Create figure legend
fig = plt.gcf()
fig.legend(handles=legend_elements, title='Board Size', loc='lower center', 
           ncol=5, fontsize=12, frameon=False, bbox_to_anchor=(0.5, -0.06))


plt.tight_layout()
plt.show()
```
